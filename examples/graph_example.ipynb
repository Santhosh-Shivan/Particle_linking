# Graph generation example

## 1. Setup
Imports needed for this tutorial

```python
import graphtrack as gt
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Create data loader

 We create the loader by calling the function `DataLoader()`. Notice we also provide augmentations to be applied to the sequence. 

```python
# Name of the dataset
DATASET = "DIC-C2DH-HeLa"

# Augmentation parameters
augmentation_list = {
    "FlipLR": {},
    "FlipUD": {},
    "FlipDiagonal": {},
}

# Seach radius for the graph edges
radius=120

# Create Loader
loader = gt.GetLoaders(DATASET, augmentation=augmentation_list, radius=radius)
```

## 3. Display graphs

The loader returns the graph information comprising the node features, edge features, and connection matrix. 

```python
graph = loader.update().resolve()
```

```python
_printer = "Number of {0}: {1}"

print(_printer.format("nodes", graph[0][0].shape[0]))
print(_printer.format("node features", graph[0][0].shape[1]) + "\n")

print(_printer.format("edges", graph[0][1].shape[0]))
print(_printer.format("edge features", graph[0][1].shape[1]) + "\n")

print(_printer.format("output nodes", graph[1][0].shape[0]))
print(_printer.format("output node features", graph[1][0].shape[1]) + "\n")

print(_printer.format("output edges", graph[1][1].shape[0]))
print(_printer.format("output edge features", graph[1][1].shape[1]) + "\n")
```

If `return_graphs` is set to `False` while resolving, the loader returns the images sequence.

```python
image, masks = loader.update().resolve(return_graphs=False)

# frame to display
frame = 7

plt.imshow(image[...,frame], cmap="gray")
plt.axis("off")
plt.show()

plt.imshow(masks[...,frame], cmap="gray")
plt.axis("off")
plt.show()
```
